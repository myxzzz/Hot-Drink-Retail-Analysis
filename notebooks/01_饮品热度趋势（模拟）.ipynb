```python
import pandas as pd
import numpy as np
import os

# 1. 自动生成一年 365 天的日历
dates = pd.date_range(start='2025-02-26', end='2026-02-26', freq='D')
df_trend = pd.DataFrame({'date': dates})

```

```python
df_trend['weekday'] = df_trend['date'].dt.dayofweek
df_trend['month'] = df_trend['date'].dt.month
```

```python
def simulate_coffee_index(row):
    base = 500

    if row['weekday'] < 5:
        base += 80
    else:
        base -= 50

    return int(base + np.random.normal(0,15))

df_trend['coffee_index'] = df_trend.apply(simulate_coffee_index, axis=1)
```

```python
def simulate_milktea_index(row):
    base = 600

    if row['month'] in [6,7,8]:
        base += 200
    elif row['month'] in [11,12,1,2]:
        base -= 50
    if row['weekday'] >= 5:
        base += 150
    return int(base + np.random.normal(0,20))

df_trend['milktea_index'] = df_trend.apply(simulate_milktea_index, axis=1)

df_trend['date'] = df_trend['date'].dt.strftime('%Y-%m-%d')
display(df_trend.head())

output_path = r"C:\Users\Administrator\Desktop\data_learn\Hot-Drink-Retail-Analysis\data\processed"
if not os.path.exists(output_path):
    os.makedirs(output_path)

df_trend.to_csv(os.path.join(output_path, "simulated_trend_1year.csv"), index=False, encoding='utf-8-sig')
print("✅ 太棒了！完美的一年期趋势底表模拟+保存成功！")
```
